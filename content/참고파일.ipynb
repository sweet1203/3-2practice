## 1. 데이터 불러오기 + 결측치 삭제

```python
import pandas as pd                  # 표(데이터프레임) 라이브러리
import seaborn as sns                # 그래프 + 데이터셋 라이브러리
import matplotlib.pyplot as plt      # 그래프 그리기 라이브러리

df = sns.load_dataset("penguins")    # 데이터 불러와 df에 저장
df = df.dropna()                     # 결측치가 있는 행 삭제
```

## 5. 산점도 — 두 수치의 관계

**예시 문제:** `bill_length_mm`(x)와 `bill_depth_mm`(y)의 **산점도**를 그리시오.

**[그래픽 요소 — 반드시 포함]**
- 제목: `Bill length vs bill depth`
- x축 라벨: `bill_length_mm` / y축 라벨: `bill_depth_mm`
- 컬러: **`tomato`**

**예시 답안 코드:**

```python
plt.scatter(df["bill_length_mm"], df["bill_depth_mm"], color="tomato", alpha=0.6)
plt.title("Bill length vs bill depth")     # 제목
plt.xlabel("bill_length_mm")               # x축 라벨
plt.ylabel("bill_depth_mm")                # y축 라벨
plt.show()                                 # 그래프 보이기
```

## 6. 히스토그램 — 한 변수의 분포

**예시 문제:** 체중(`body_mass_g`)의 **히스토그램**을 그리시오.

**[그래픽 요소 — 반드시 포함]**
- 제목: `Body mass distribution`
- x축 라벨: `body_mass_g` / y축 라벨: `count`
- 컬러: **`skyblue`**

**예시 답안 코드:**

```python
sns.histplot(data=df, x="body_mass_g", bins=15, color="skyblue")
plt.title("Body mass distribution")        # 제목
plt.xlabel("body_mass_g")                  # x축 라벨
plt.ylabel("count")                        # y축 라벨
plt.show()                                 # 그래프 보이기
```

## 7. 그룹별 평균 막대그래프 — groupby

**예시 문제:** 종(`species`)별 **평균 체중(`body_mass_g`)**을 **막대그래프**로 그리시오.

**[그래픽 요소 — 반드시 포함]**
- 제목: `Mean body mass by species`
- x축 라벨: `species` / y축 라벨: `mean body_mass_g`
- 컬러: **`steelblue`**

**예시 답안 코드:**

```python
df.groupby("species")["body_mass_g"].mean().plot(kind="bar", color="steelblue")
plt.title("Mean body mass by species")     # 제목
plt.xlabel("species")                      # x축 라벨
plt.ylabel("mean body_mass_g")             # y축 라벨
plt.show()                                 # 그래프 보이기
```
